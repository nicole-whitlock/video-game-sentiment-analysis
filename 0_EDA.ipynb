```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from langdetect import detect
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('games_reviews.csv')
```

```python
df.head()
```

Detect la

```python
def det(x):
    try:
        lang = detect(x)
    except:
        lang = 'Other'
    return lang

df['Lang'] = df['quote'].apply(det)
```

```python
df.head()
```

```python
eng_df = df[df['Lang'] == 'en'].reset_index(drop=True)
```

```python
eng_df.head()
```

Keep only the quote and score columns, those are the only ones being used

```python
df = eng_df[['quote', 'score']].copy()
```

Check for null values

```python
df.shape
```

```python
df.isnull().sum()
```

Drop null score values, it's a very small percentage of the total number in the dataset

```python
df.dropna(inplace=True)
```

Show the min and max of score

```python
print(df.score.min())
print(df.score.max())
```

Show distribution of word counts in quote

```python
df['words'] = [len(quote.split()) for quote in df['quote']]
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['words'], kde=True, bins=20)
plt.title('Distribution of Word Counts in Quotes')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Group Score into 'positive', 'negative', and 'neutral' sentiment




```python
df['sentiment'] = ['positive' if score > 70 else 'negative' if score < 40 else 'neutral' for score in df['score']]
```

```python
df.sentiment.value_counts()
```

Plot the distribution of words counts

```python
df['sentiment'].value_counts().plot(kind='bar', title='Sentiment Counts')
```

Save dataset

```python
df.to_csv('games_reviews_eng.csv', index=False)
```
